```python
# Marketing_course_2.1 #########################################
import pandas as pd
import numpy as np
df = pd.read_csv('marketing_data.csv')
```

```python
#################################################################
# Split DF on two numeric and categorical
def splitdf(df):
    numeric_df = df.select_dtypes(include=["number"])      # numeric_df (int64 or float64) 
    categorical_df = df.select_dtypes(exclude=["number"])  # objects
    return numeric_df, categorical_df
```

```python
#################################################################
def summary(df_csv):
    # Create a summary DataFrame
    summary_df = pd.DataFrame({
        "Data Type": df_csv.dtypes,
        "Memory Used": df_csv.memory_usage(deep=True),
        "Nunique": df_csv.nunique(),
        "Max": df_csv.select_dtypes(include='number').max(),
        "Min": df_csv.select_dtypes(include='number').min(),
        "Missing": df_csv.isnull().sum(),
        'Whitespace': df_csv.map(lambda x: isinstance(x, str) and x.strip() == '').sum()
    })

    summary_df["Min"] = summary_df["Min"].fillna("N/A")
    summary_df["Max"] = summary_df["Max"].fillna("N/A")
    return summary_df

#################################################################
summary(df)
```

```python
# Delete all white spaces (leading and trailing) from the names of the columns of DF 
# Target 'Income' column
df.columns = df.columns.str.strip()
```

```python
# rename Multyple columns in DF (in place)
df = df.rename(columns= {
'ID': 'Cust_ID',
'Year_Birth': 'Cust_Year_Birth',
'Education': 'Cust_Education',
'Marital_Status': 'Cust_Marital_Status',
'Income': 'Cust_Income',
'Kidhome': 'Cust_Kid',
'Teenhome': 'Cust_Teen',
'Dt_Customer': 'Enrollment_Date',
'Recency': 'Days_Since_Last_Purchase',
'MntWines': '2Y_Amount_Wines',
'MntFruits': '2Y_Amount_Fruits',
'MntMeatProducts': '2Y_Amount_Meat',
'MntFishProducts': '2Y_Amount_Fish',
'MntSweetProducts': '2Y_Amount_Sweet',
'MntGoldProds': '2Y_Amount_Gold',
'NumDealsPurchases': 'N_Discount_Purchase',
'NumWebPurchases': 'N_Web_Purchase',
'NumCatalogPurchases': 'N_Catalogue_Purchase',
'NumStorePurchases': 'N_Store_Purchase',
'NumWebVisitsMonth': 'N_lastMnth_webVisits',
'AcceptedCmp3': 'Offer_Accepted_3',
'AcceptedCmp4': 'Offer_Accepted_4',
'AcceptedCmp5': 'Offer_Accepted_5',
'AcceptedCmp1': 'Offer_Accepted_1',
'AcceptedCmp2': 'Offer_Accepted_2',
'Response': 'Offer_Accepted_Last',
'Complain': 'Complain_2Y',
'Country': 'Cust_Location'
})
```

```python
# Check for duplicates
has_duplicates = df.duplicated().any()
print(f"\n Has duplicates: {has_duplicates}")
df.duplicated().sum()
```

```python
# Remove '$', commas, and whitespace, then convert to float for values in the 'Income' column
df['Cust_Income'] = pd.to_numeric(
    df['Cust_Income'].str.replace(r'[\$, ]', '', regex=True),
    errors='coerce'  # Ensures invalid entries (like missing data) become NaN instead of crashing
)
```

```python
# Split DF into Numeric and Categorical
numeric_df, categorical_df = splitdf(df)
```

```python
# Found column and rows where at least one value is negative
negative_values = numeric_df.columns[(numeric_df < 0).any()]
print("\n # 1.2 -------------- Columns with values less than 0:\n", negative_values)
```

```python
# Work with Missing 'Income' values
# 1.Consider that customers with similar education and marital status tend to have comparable yearly incomes, on average
rows_no_missing = df.dropna()   # Create df with no NaN values to find avrg Income for combination of 'Marital_Status' and 'Education'
rows_no_missing.isnull().sum()
```

```python
rows_no_missing.isnull().any()
```

```python
# 2. Get an avrg Income for combination of 'Marital_Status' and 'Education' to fill NaN in 'Income' 
# 2.1. Compute average Income per group
group_avg_df = rows_no_missing.groupby(['Cust_Marital_Status', 'Cust_Education'])['Cust_Income'].median().round()
'''
.reset_index(name='Average_Income')   # After reset_index(name='Average_Income'), group_avg_df no longer has a MultiIndex, so 
we can’t use tuples like (row['Cust_Marital_Status'], row['Cust_Education']) to index rows.

'''

group_avg_df
```

```python
# 2.2 Fill NaNs with the avarege avrg Income for combination of 'Marital_Status' and 'Education'
df['Cust_Income'] = df.apply(
    lambda row: group_avg_df.loc[(row['Cust_Marital_Status'], row['Cust_Education'])] 
    if pd.isna(row['Cust_Income']) else row['Cust_Income'],
    axis=1
)
```
